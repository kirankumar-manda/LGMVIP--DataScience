### Loading Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing Dataset

```python
df = pd.read_csv("Terrorism.csv")
```

Shape of the Dataset

```python
df.shape
```

```python
df.head()
```

Top 10 Countries affected by terrorism.

```python
print("Most Country have a Terrorism is: ", df["country_txt"].value_counts().index[0])
print("Most Region have a Terrorism is: ", df["region_txt"].value_counts().index[0])
print("The most years which terrorism occurred was among: ", df["iyear"].value_counts().index[:5].sort_values())
```

### Terrorism Activities Each Year

```python
plt.subplots(figsize=(16,6))
sns.countplot(df['iyear'], data=df, palette='flare')#RdBu
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("Year")
plt.title('Terrorism Activities Each Year')
plt.show()
```

```python
df_India=df[df["country_txt"] == "India"]
```

```python
df_India.head()
```

### Terrorism Activities Each Year in India

```python
plt.subplots(figsize=(15,6))
sns.countplot(df_India['iyear'], data=df, palette='flare')
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("Year")
plt.title('Terrorism Activities Each Year')
plt.show()
```

### Number Of Terrorist Activities By Region

```python
plt.subplots(figsize=(10,6))
sns.countplot('region_txt',data=df,palette='RdYlGn',edgecolor=sns.color_palette('dark',4),order= df['region_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel("Regions")
plt.title('Number Of Terrorist Activities By Region')
plt.show()
```

### Number of Terrorist Targets in India

```python
plt.subplots(figsize=(12,6))
sns.countplot(df_India['targtype1_txt'], data=df, palette='RdBu',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.xlabel("Target Area")
plt.title('Number Of Terrorist Targets')
plt.show()
```

### Successful & Failure Operation in India

```python
data_copy = df_India.copy()
data_copy["success"] = df_India["success"].replace({1:"Successful Operation", 0:"Failure Operation"})
data_copy["success"].value_counts().plot(kind="pie", autopct="%1.1f%%", cmap="Accent")
plt.title("Successful & Failure Operation")
```

### Most Countries Affected By Terrorism

```python
plt.subplots(figsize=(15,6))
sns.countplot('country_txt',data=df,palette='inferno',order=df["country_txt"].value_counts().index[:10])
plt.xticks(rotation=90)
plt.title('Most Countries Affected By Terrorism')
plt.xlabel("Country")
plt.grid(axis="y")
plt.show()
```

### The Most Active Terror Groups

```python

sns.barplot(df['gname'].value_counts()[:5].values,df['gname'].value_counts()[:5].index)#,palette=('cividis_r'))
plt.xticks(rotation=0)
fig=plt.gcf()
fig.set_size_inches(20,10)
plt.title('The Most Active Terror Groups')
plt.show()
```
